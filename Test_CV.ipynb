```python
# Импортирую нужные библиотеки

import pandas as pd
import numpy as np
```

```python
# Загружаю файл с данными 

db = pd.read_csv('Test.csv', sep = ';')
```

```python
db.head(22)
```

```python
# Формирую датасет с фильмами и рейтингом. Самый высокий рейтинг у фильма с наибольшим числом просмотров.

db_new = db['Movies'].value_counts().reset_index()
db_new.columns = ['Movies','Rating']
```

```python
db_new.head()
```

```python
s = db_new.loc[1]
```

```python
# Объединяю исходный датафрейм и датафрейм с рейтингом фильма

df = db.merge(db_new, on = 'Movies', how = 'left')
df.head(22)
```

```python
# Функция для выбора фильма. 
# Если пользователь посмотрел фильм с максимальным рейтингом, то ему будет предложен фильм с рейтингом меньшим на 1.
# Если пользователь не смотрел фильм с максимальным рейтингом, то ему будет предложен этот фильм. 

def max_rate_2 (param):
    max_ind = df['Rating'].argmax()
    if (param != df['Movies'][max_ind]):
        return(df['Movies'][max_ind])  
    if (param == df['Movies'][max_ind]):     
        return s[0]   
```

```python
# Применяю функцию, записываю в новый столбец Recommendation

df['Recommendation'] = df['Movies'].apply(max_rate_2)
```
