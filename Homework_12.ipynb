Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
ratings.head(1000)
```

```python
### датафрейм оформляем в сводную таблицу, где строки - userID, столбцы -  количество оценок (len), max (timestamp), 
### min(timestamp)

ratings1 = ratings.pivot_table(index = 'userId', values = 'timestamp', aggfunc = [len, max, min])
ratings1.head()
```

```python
### отбираем тех пользователей, которые выставили более 100 оценок

ratings2 = ratings1.loc[ratings1['len']['timestamp'] > 100]
ratings2.head()
```

```python
### считаем время жизни = макс-мин

ratings2['age'] = ratings2['max']['timestamp'] - ratings2['min']['timestamp']
ratings2.head()
```

```python
### вычисляем среднее время жизни

print('Cреднее время жизни пользователей, которые выставили более 100 оценок:', round(ratings2['age'].mean()))
```
